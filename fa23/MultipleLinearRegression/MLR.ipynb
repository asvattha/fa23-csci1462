```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
housing = pd.read_csv('data/housing.csv')
housing.head()
```

```python
use_cols = ['housing_median_age', 'total_rooms', 'median_income', 'median_house_value'] 
use_cols
```

```python
housing = housing[use_cols]
housing.head()
```

```python
from pandas.plotting import scatter_matrix

_ = scatter_matrix(housing[use_cols], figsize=(12,8))
```

##  Base line model (Constant model)

```python
pred_house_value = np.mean(housing['median_house_value'])

# note we are considering the entire data as train data, but in reality we first divide the data into train and test and then compute
# model parameter on the train data. 
```

```python
rmse = np.sqrt(np.sum((pred_house_value - housing['median_house_value'])**2))
rmse
```
